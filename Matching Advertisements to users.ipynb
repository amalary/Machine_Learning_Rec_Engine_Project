```python
import pandas as pd 
from sklearn.naive_bayes import GaussianNB -
```

```python
df =  pd.read_csv('~/Downloads/archive/advertising.csv')

df.head()
```

```python
del df['Ad Topic Line']
del df['Timestamp']
del df['City']
del df['Country']
del df['Male']
```

```python
df
```

```python
# Assign all columns in the dataset as X variables, excluding "Clicked Ads"

X = df.drop('Clicked on Ad', axis = 1)

# Assign "Clicked on Ad" as y variable 

y  = df['Clicked on Ad']
```

```python
model = GaussianNB()
```

```python
model.fit(X,y)
```

```python
target_user =[


    66.00, # Daily Time Spent on Site 
    48, #Age 
    24593.33, #Area Income 
    131.76, #Daily Internet Usage 
]

pred_target_user = model.predict([target_user])
print("Prediction:", pred_target_user)
```
