```python
import pandas as pd 
from sklearn.neighbors import NearestNeighbors 

df = pd.read_csv('~/Downloads/archive/Melbourne_housing_FULL.csv')

df.head(10)
```

```python
df.dropna(axis = 0, how = 'any',subset = None, inplace = True)
```

```python
df.head()
```

```python
X = df.loc[:, ['Price', 'Distance','Bedroom2', 'Bathroom','Landsize', 'BuildingArea','YearBuilt']].values
```

```python
target = [1350000,2,2,2,220,200,2005]
```

```python
model = NearestNeighbors(n_neighbors = 3).fit(X)
```

```python
print(model.kneighbors([target]))
```
